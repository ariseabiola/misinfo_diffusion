````python
import pymongo
import os
from dotenv import find_dotenv, load_dotenv

from src.utils import get_topics_in_db
from src.visualization.visualize import create_collection_figure_trace
import plotly.graph_objs as go
import plotly.offline as pyo

pyo.init_notebook_mode()

load_dotenv(find_dotenv())
````

````python
db_name = os.environ.get('DB_NAME')
host_ = 'localhost'
port = int(os.environ.get('PORT'))
client = pymongo.MongoClient(host=host_, port=port)

db = client[db_name]
````

````python
def see_all_topics_in_db(**kwargs):
    return list(get_topics_in_db(db=kwargs['db']).keys())
````

See all topics in your database

````python
see_all_topics_in_db(db=db)
````

#### Draw Plot

To draw plot for one topic:

```code
traces, title_text = create_collection_figure_trace('trump-hillary', db=db)
```

To draw plot for more than one topic (e.g. trump-hillary, football, music):
```code
traces, title_text = create_collection_figure_trace('trump-hillary', 'football', 'music', db=db)
```

````python
traces, title_text = create_collection_figure_trace('trump-hillary', db=db)
````

````python
fig = go.Figure(traces)

fig.update_layout(title_text=title_text,
                  xaxis_rangeslider_visible=True)

pyo.iplot(fig)
````
